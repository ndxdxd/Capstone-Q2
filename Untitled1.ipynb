```python
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import cv2
import json
import random

# Load the model
resnet50 = ResNet50(weights='imagenet', include_top = True, classifier_activation= None)
EPS = 0.5 / 255  # Reduced perturbation magnitude

# Load ImageNet labels
IMAGENET_LABELS = "./data/imagenet_class_index.json"
with open(IMAGENET_LABELS) as f:
    IMAGENET_CLASSES = {int(i): x[1] for i, x in json.load(f).items()}

# Load and display image
def show_image(image_path):
    image_pixels = plt.imread(image_path)
    plt.imshow(image_pixels)
    plt.show()
    return image_pixels

# Preprocess image for model input
def preprocess_image(image_pixels):
    image_pixels = cv2.resize(image_pixels, (224, 224))
    image_pixels = np.expand_dims(image_pixels, axis=0)
    return image_pixels

# Clip perturbation to valid range
def clip_eps(delta_tensor):
    return tf.clip_by_value(delta_tensor, clip_value_min=-EPS, clip_value_max=EPS)

# Generate targeted adversarial perturbation
def generate_adversaries_targeted(image_tensor, delta, model, true_index, target_index):
    loss_fn = tf.keras.losses.SparseCategoricalCrossentropy()
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)  # Reduced learning rate
    
    for t in range(100):  # Reduced number of iterations
        with tf.GradientTape() as tape:
            tape.watch(delta)
            inp = preprocess_input(image_tensor + delta)
            predictions = model(inp, training=False)
            loss = (-loss_fn([true_index], predictions) + loss_fn([target_index], predictions))
        
        gradients = tape.gradient(loss, delta)
        optimizer.apply_gradients([(gradients, delta)])
        delta.assign_add(clip_eps(delta))
    
    return delta

# Apply multiple perturbations to the same image and compare logits
def perturb_image(image_path, true_label, model, k=5):
    sample_image = show_image(image_path)
    preprocessed_image = preprocess_image(sample_image)
    
    image_tensor = tf.constant(preprocessed_image, dtype=tf.float32)
    delta = tf.Variable(tf.zeros_like(image_tensor), trainable=True)
    
    # Randomly sample k target indices
    target_labels = random.sample(list(IMAGENET_CLASSES.keys()), k)
    
    # Get logits before perturbations
    original_logits = model.predict(preprocess_input(image_tensor))
    print("Logits before perturbations:")
    for target in target_labels:
        print(f"Target {IMAGENET_CLASSES[target]}: {original_logits[0][target]}")
    
    for target in target_labels:
        print(f"Applying perturbation for target: {IMAGENET_CLASSES[target]}")
        delta = generate_adversaries_targeted(image_tensor, delta, model, true_label, target)
    
    perturbed_image = preprocess_input(image_tensor + delta)
    final_preds = model.predict(perturbed_image)
    
    # Get logits after perturbations
    print("Logits after perturbations:")
    for target in target_labels:
        print(f"Target {IMAGENET_CLASSES[target]}: {final_preds[0][target]}")
    
    print("Final Predictions:", decode_predictions(final_preds, top=3)[0])
    
    plt.imshow((image_tensor + delta).numpy().squeeze()/255)
    plt.show()

# Example usage
perturb_image("pig.jpg", 341, resnet50, k=5)
```
